````python
x = [] # the position
y = [] # the moves
def race_to_finish(target):
    q = deque([(0,1,0)])
    x.append(0)
    y.append(0)
    while len(q)>0:
        v = q.popleft()
        (moves, speed, pos) = v
        if pos==target:
            return moves
        q.append((moves+1, 2*speed, pos+speed))
        y.append(moves+1)
        x.append(pos+speed)
        rspeed = -1 if speed>0 else 1
        if (pos+speed>target and speed>0) or (pos+speed<target and speed<0):
            y.append(moves+1)
            x.append(pos)
            q.append((moves+1,rspeed, pos))
    return -1
````

````python
race_to_finish(10)
````

````python
plt.figure(figsize=(20,10))
plt.scatter(x,y)
plt.xlabel("position")
plt.ylabel("moves")
plt.xlim(-75,100)
plt.xticks(range(-10,260,10))
plt.show()
````

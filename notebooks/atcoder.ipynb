```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2, 5, 0, 3, 3, 1])
y = np.array([0, 2, 3, 2, 4, 5])
s = 3

plt.figure(figsize=(10, 3))  # 10 inches wide, 4 inches tall

# Create the first plot
plt.subplot(1, 3, 1)  # 1 row, 2 columns, plot number 1
plt.plot(x, np.zeros(x.shape), color = "k")
plt.title('x-axis')

# Create the second plot
plt.subplot(1, 3, 2)  # 1 row, 2 columns, plot number 2
plt.plot(np.zeros(y.shape), y, color = "k")
plt.title('y-axis')

plt.subplot(1, 3, 3) # 1 row, 3 columns, plot number 3
plt.scatter(x, y, color = "k")
plt.title('xy plane')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
center_points = np.arange(0, 20, 1)
def cost(points, cp):
    res = 0
    for p in points:
        if p < cp - s:
            res += cp - s - p
        elif p > cp + s:
            res += p - cp - s
    return res
cost_x = np.array([cost(x, cp) for cp in center_points])
cost_y = np.array([cost(y, cp) for cp in center_points])

# Create the first plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot number 1
plt.scatter(center_points, cost_x, color = "k")
plt.title('x cost')

# Create the second plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot number 2
plt.scatter(center_points, cost_y, color = "k")
plt.title('y cost')


plt.show()
```

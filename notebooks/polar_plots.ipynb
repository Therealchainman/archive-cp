```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an array of angles from 0 to 2*pi
theta = np.linspace(0, 100*np.pi, 30_000)

# Compute the complex numbers using e^(ix)
# r = np.exp(1j * theta) + np.exp(1j * theta * np.pi)
r = np.exp(1j * np.pi * theta)
# Plot in polar coordinates
plt.figure(figsize=(8, 8))
plt.polar(theta, r, label='r = sin(2*theta)')

plt.title('Polar Plot Example')
plt.legend()
plt.show()
```
